Generating images with AI, especially with smaller models intended for CPU, might still be slow depending on your system's capabilities. Results may vary.

```python
# Install necessary libraries
%pip install diffusers transformers scipy accelerate

# Import libraries
import torch
from diffusers import StableDiffusionPipeline

# Check if MPS is available (for M1/M2 Macs) or fallback to CPU
device = "cpu" # We are targeting CPU

# Load a smaller model suitable for CPU. Note: Even smaller models can be resource intensive.
# This is a good starting point, you might need to experiment with other small models.
model_id = "runwayml/stable-diffusion-v1-5" # A relatively common and widely supported model

# Load the pipeline
# Use torch_dtype=torch.float32 when running on CPU for potentially better compatibility
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)
pipe = pipe.to(device)

# Generate an image
prompt = "a photo of an astronaut riding a horse on the moon" # You can change this prompt
print("Generating image...")
image = pipe(prompt).images[0]

# Display the image
display(image)

print("Image generation finished.")
```
